## Imports

```python
import lightgbm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn.ensemble
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import ConfusionMatrixDisplay, f1_score
from torchgeo.datasets import QuakeSet
from tqdm.auto import tqdm
from itertools import product
```

## Datasets and features import
Those are produced in the last section of the main notebook.

```python
train_set = QuakeSet(split="train")
val_set = QuakeSet(split="val")
test_set = QuakeSet(split="test")
```

```python
train, val, test = pd.read_csv("train.csv"), pd.read_csv("val.csv"), pd.read_csv("test.csv")
complete = pd.concat([train, val, test])
```

## Validity of the zero-difference hypothesis
We show that on the whole train+val+test set, only one item has a zero difference despite the fact that there is an earthquake. Indeed, this item is all-zero, despite the fact that it is labelled positive and has a magnitude of 5.49. This item is in the training set and can be discarded.

Aside from this outlier, the zero-difference rule is correct in all the cases.

```python
outlier = complete[(complete["zero_diff"]==1) & (complete["label"]==1)].index[0]
print(outlier)
train_set.plot(train_set[outlier])
```

## Robustness
We compute the different metrics when the model is trained and validated on different portions of the dataset. We show that our model is pretty robust to the different combinations.
### Classification robustness
#### Inverse learning

```python
sets = ["train.csv", "test.csv", "val.csv"]
```

```python
def evaluate(train, val):
    val_zero_differences_indices = val[val["zero_diff"]==1].index
    train_drop = train[train["zero_diff"]==1].index
    
    classifier = lightgbm.LGBMClassifier(verbose=-1, random_state=42).fit(
        train.drop(["label", "magnitude"], axis=1).drop(train_drop, axis=0), 
        train["label"].drop(train_drop, axis=0))
    val_classifications = classifier.predict(val.drop(["label", "magnitude"], axis=1)).astype(np.int8)
    val_classifications[val_zero_differences_indices] = 0
    score = f1_score(val["label"], np.round(val_classifications))
    return score
```

```python
classification_scores = {}
for i, j in product(sets, sets):
    if i != j:
        classification_scores["X: "+str(i)[:-4]+", "+"x: "+str(j)[:-4]] = (evaluate(pd.read_csv(i), pd.read_csv(j)))
print("Mean F1-score:", np.mean(list(classification_scores.values())))
print("Std F1-score:", np.std(list(classification_scores.values())))
```

```python
sns.set_theme(rc={'figure.figsize':(12, 5)}, style="whitegrid")
sns.barplot(classification_scores)
plt.title("Classification F1-score")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.savefig("f1_0.png")
plt.show()
```

#### Combinations of datasets

```python
classification_scores = {}
for i, j, k in product(sets, sets, sets):
    if i!=j and i!=k and j!=k:
        classification_scores["X: "+str(i)[:-4]+"+"+str(j)[:-4]+" x: "+str(k)[:-4]] = (evaluate(pd.concat([pd.read_csv(i), pd.read_csv(j)]), pd.read_csv(k)))
print("Mean F1-score:", np.mean(list(classification_scores.values())))
print("Std F1-score:", np.std(list(classification_scores.values())))
```

```python
sns.set_theme(rc={'figure.figsize':(12, 5)}, style="whitegrid")
sns.barplot(classification_scores)
plt.title("Classification F1-score")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.savefig("f1_1.png")
plt.show()
```

### Regression robustness
#### Inverse learning

```python
def evaluate(train, val):
    val_zero_differences_indices = val[val["zero_diff"]==1].index
    unaffected = train[train["magnitude"]==0].index
    unaffected = unaffected[:len(unaffected)//2]
    
    regressor = lightgbm.LGBMRegressor(verbose=-1, random_state=42).fit(
        train.drop(["magnitude", "zero_diff"], axis=1).drop(unaffected, axis=0), 
        train["magnitude"].drop(unaffected, axis=0)
    )
    val_regression = regressor.predict(val.drop(["magnitude", "zero_diff"], axis=1))
    val_regression[val_zero_differences_indices] = 0
    val_regression[val_regression < 4] = 0
    val_regression[val_regression > 10] = 10
    score = np.mean(np.abs(val["magnitude"]-val_regression))
    return score
```

```python
regression_scores = {}
for i, j in product(sets, sets):
    if i != j:
        regression_scores["X: "+str(i)[:-4]+", "+"x: "+str(j)[:-4]] = (evaluate(pd.read_csv(i), pd.read_csv(j)))
print("Mean MAE:", np.mean(list(regression_scores.values())))
print("Std MAE:", np.std(list(regression_scores.values())))
```

```python
sns.set_theme(rc={'figure.figsize':(12, 5)}, style="whitegrid")
sns.barplot(regression_scores)
plt.title("Regression MAE")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.savefig("mae_0.png")
plt.show()
```

#### Combination of dataset

```python
regression_scores = {}
for i, j, k in product(sets, sets, sets):
    if i!=j and i!=k and j!=k:
        regression_scores["X: "+str(i)[:-4]+"+"+str(j)[:-4]+" x: "+str(k)[:-4]] = (evaluate(pd.concat([pd.read_csv(i), pd.read_csv(j)]), pd.read_csv(k)))
print("Mean MAE:", np.mean(list(regression_scores.values())))
print("Std MAE:", np.std(list(regression_scores.values())))
```

```python
sns.set_theme(rc={'figure.figsize':(12, 5)}, style="whitegrid")
sns.barplot(regression_scores)
plt.title("Regression MAE")
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.savefig("mae_1.png")
plt.show()
```
